```python
from twitter import *
```

```python
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

t = Twitter(auth=OAuth(access_token,access_token_secret, consumer_key, consumer_secret))
```

```python
import os
output_filename = os.path.join(os.path.expanduser("~"),
"Desktop","Graph-Mining-on-Twitter" ,"rajinikanth_tweets.json")
```

```python
import json
```

```python
with open(output_filename, 'a') as output_file:
    search_results = t.search.tweets(q="rajinikanth", count=100)['statuses']
    for tweet in search_results:
        if 'text' in tweet:
            print(tweet['user']['screen_name'])
            print(tweet['text'])
            print()
            output_file.write(json.dumps(tweet))
            output_file.write("\n\n")
```
